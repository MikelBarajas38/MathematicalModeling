```python
def plot_dicesum_hist(n, c, d):
    
    count = {}
    res = []

    for i in range(n):

        sum = 0

        for j in range(c):
            dice_roll = int(random.random() * d) + 1
            sum += dice_roll

        if sum not in count:
            count[sum] = 0

        count[sum] = count[sum] + 1
        res.append(sum)

    plt.hist(res, align='mid')
    plt.show()


for i in range(1, 6):
    plot_dicesum_hist(10000, i, 6)
```
